```python
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```python
text = "Artificial Intelligence and Machine Learning are transforming industries, making systems smarter and more efficient!"
```

```python
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()
print("Whitespace Tokenization:")
print(wt.tokenize(text))
```

```python
from nltk.tokenize import wordpunct_tokenize

print("\nPunctuation-based Tokenization:")
print(wordpunct_tokenize(text))
```

```python
from nltk.tokenize import TreebankWordTokenizer

twt = TreebankWordTokenizer()
print("\nTreebank Tokenization:")
print(twt.tokenize(text))
```

```python
from nltk.tokenize import TweetTokenizer

tt = TweetTokenizer()
print("\nTweet Tokenization:")
print(tt.tokenize(text))
```

```python
from nltk.tokenize import MWETokenizer

mwe = MWETokenizer([('Artificial', 'Intelligence'), ('Machine', 'Learning')])
tokens = text.split()

print("\nMWE Tokenization:")
print(mwe.tokenize(tokens))
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer

words = ["transforming", "industries", "making", "efficient"]

ps = PorterStemmer()
ss = SnowballStemmer("english")

print("\nPorter Stemming:")
for w in words:
    print(w, "->", ps.stem(w))

print("\nSnowball Stemming:")
for w in words:
    print(w, "->", ss.stem(w))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print("\nLemmatization:")
for w in words:
    print(w, "->", lemmatizer.lemmatize(w))
```
