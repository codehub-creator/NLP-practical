```python
# Morphology Practical with Add and Delete Examples

import pandas as pd
from IPython.display import display

# Data with add/delete operations
data = {
    "Root Word": ["play", "teach", "happy", "make", "use", "write"],
    "Add Morpheme": ["-ing", "-er", "un-", "-ing", "-less", "re-"],
    "Delete Letter": ["None", "None", "None", "e", "None", "None"],
    "New Word": ["playing", "teacher", "unhappy", "making", "useless", "rewrite"],
    "Meaning": [
        "Present continuous action",
        "Person who teaches",
        "Not happy",
        "Action in progress",
        "Without use",
        "Write again"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Style Table
styled_table = df.style.set_caption("Morphology Add-Delete Table") \
    .set_table_styles([
        {'selector': 'caption',
         'props': [('font-size', '18px'),
                   ('font-weight', 'bold'),
                   ('text-align', 'center'),
                   ('padding', '10px')]},
        {'selector': 'th',
         'props': [('background-color', '#2E75B6'),
                   ('color', 'white'),
                   ('text-align', 'center')]},
        {'selector': 'td',
         'props': [('text-align', 'center'),
                   ('padding', '8px')]},
        {'selector': 'table',
         'props': [('border-collapse', 'collapse'),
                   ('width', '100%')]}
    ])

display(styled_table)
```
